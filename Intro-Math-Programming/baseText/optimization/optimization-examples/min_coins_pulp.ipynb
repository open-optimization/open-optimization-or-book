```python
# Minimum coins for 83 cents — PuLP
import pulp as pl
prob = pl.LpProblem("Min_Coins", pl.LpMinimize)
p = pl.LpVariable("pennies",  lowBound=0, cat="Integer")
n = pl.LpVariable("nickels",  lowBound=0, cat="Integer")
d = pl.LpVariable("dimes",    lowBound=0, cat="Integer")
q = pl.LpVariable("quarters", lowBound=0, cat="Integer")
prob += p + n + d + q, "Total_coins"
prob += p + 5*n + 10*d + 25*q == 83, "Make_83_cents"
prob.solve()
print("Status:", pl.LpStatus[prob.status])
print(f"pennies={p.value():g}, nickels={n.value():g}, dimes={d.value():g}, quarters={q.value():g}")
print("Total coins =", pl.value(prob.objective))
```
