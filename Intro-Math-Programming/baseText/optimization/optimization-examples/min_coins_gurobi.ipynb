```python
# Minimum coins for 83 cents — gurobipy
import gurobipy as gp
from gurobipy import GRB
m=gp.Model("Min_Coins")
p=m.addVar(vtype=GRB.INTEGER,lb=0,name="pennies"); n=m.addVar(vtype=GRB.INTEGER,lb=0,name="nickels")
d=m.addVar(vtype=GRB.INTEGER,lb=0,name="dimes"); q=m.addVar(vtype=GRB.INTEGER,lb=0,name="quarters")
m.setObjective(p+n+d+q,GRB.MINIMIZE)
m.addConstr(p+5*n+10*d+25*q==83,"Make_83_cents")
m.optimize()
if m.status==GRB.OPTIMAL:
    print(f"pennies={p.X:g}, nickels={n.X:g}, dimes={d.X:g}, quarters={q.X:g}"); print("Total coins =",m.ObjVal)
```
