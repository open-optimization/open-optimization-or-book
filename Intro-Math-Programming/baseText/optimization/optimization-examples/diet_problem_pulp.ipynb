```python
# Solving the Diet Problem using PuLP

import pulp as pl

# Define the problem
prob = pl.LpProblem("Diet_Problem", pl.LpMinimize)

# Decision Variables
x1 = pl.LpVariable("green", 0, None, pl.LpInteger)
x2 = pl.LpVariable("blue", 0, None, pl.LpInteger)
x3 = pl.LpVariable("yellow", 0, None, pl.LpInteger)
x4 = pl.LpVariable("red", 0, None, pl.LpInteger)

# Objective Function
prob += 1.25 * x1 + 1.05 * x2 + 0.85 * x3 + 0.65 * x4, "Total Cost"

# Constraints
prob += 6 * x1 + 4 * x2 + 5 * x3 + 3 * x4 >= 20, "Iron"
prob += 6 * x1 + 5 * x2 + 2 * x3 + 6 * x4 >= 25, "Vitamin B"
prob += 7 * x1 + 4 * x2 + 5 * x3 + 3 * x4 >= 30, "Vitamin C"
prob += 4 * x1 + 9 * x2 + 6 * x3 + 2 * x4 >= 15, "Vitamin D"

# Solve the problem
prob.solve()

# Print the results
print("Status:", pl.LpStatus[prob.status])
for v in prob.variables():
    print(v.name, "=", v.varValue)
```
