```python
# Airline Transfer Network: Maximum Flow s -> t — gurobipy
import gurobipy as gp
from gurobipy import GRB
cap = {("s","a"):4,("s","b"):2,("s","c"):8,("a","c"):5,("a","d"):2,("b","c"):6,
       ("b","e"):9,("c","d"):1,("c","e"):3,("c","t"):4,("d","e"):7,("e","t"):5}
V = {"s","a","b","c","d","e","t"}; s,t="s","t"
m = gp.Model("Max_Flow")
x = {a: m.addVar(lb=0, ub=u, name=f"x_{a[0]}_{a[1]}") for a,u in cap.items()}
m.setObjective(gp.quicksum(x[a] for a in cap if a[0]==s), GRB.MAXIMIZE)
for k in V-{s,t}:
    m.addConstr(gp.quicksum(x[a] for a in cap if a[0]==k)==gp.quicksum(x[a] for a in cap if a[1]==k), f"Balance_{k}")
m.optimize()
if m.status==GRB.OPTIMAL:
    print("Max flow =", m.ObjVal)
    for a in cap: print(f"{a[0]}->{a[1]}: {x[a].X:g}")
```
