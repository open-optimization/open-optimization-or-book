```python
# Assignment Problem (Hiring for tasks) — gurobipy
import gurobipy as gp
from gurobipy import GRB

cost = {(1,1):40,(1,2):47,(1,3):80,
        (2,1):72,(2,2):36,(2,3):58,
        (3,1):24,(3,2):61,(3,3):71}
P = [1,2,3]; T = [1,2,3]

m = gp.Model("Assignment")
x = m.addVars(cost.keys(), vtype=GRB.BINARY, name="x")
m.setObjective(gp.quicksum(cost[i,j]*x[i,j] for i,j in cost), GRB.MINIMIZE)
m.addConstrs((gp.quicksum(x[i,j] for j in T) == 1 for i in P), "Person")
m.addConstrs((gp.quicksum(x[i,j] for i in P) == 1 for j in T), "Task")

m.optimize()
if m.status == GRB.OPTIMAL:
    for (i,j),v in x.items():
        if v.X > 0.5:
            print(f"Person {i} -> Task {j}  (cost {cost[i,j]})")
    print("Total cost = $", m.ObjVal)
```
