```python
# Bakery Production — PuLP
# Decision: number of cakes (c), cookies (k), muffins (m) to bake per day.
# max 6 c + 1.5 k + 2.5 m
# flour: 500 c + 100 k + 200 m <= 4800
# sugar: 200 c +  50 k +  80 m <= 2060
# time : 60 c +  10 k +  20 m <= 480
import pulp as pl

prob = pl.LpProblem("Bakery_Production", pl.LpMaximize)

# Use integer variables (whole baked goods); drop cat=Integer for the LP relaxation.
c = pl.LpVariable("cakes",   lowBound=0, cat="Integer")
k = pl.LpVariable("cookies", lowBound=0, cat="Integer")
mu = pl.LpVariable("muffins", lowBound=0, cat="Integer")

prob += 6*c + 1.5*k + 2.5*mu, "Total_Profit"
prob += 500*c + 100*k + 200*mu <= 4800, "Flour_g"
prob += 200*c +  50*k +  80*mu <= 2060, "Sugar_g"
prob +=  60*c +  10*k +  20*mu <= 480,  "Baking_minutes"

prob.solve()
print("Status:", pl.LpStatus[prob.status])
for v in prob.variables():
    print(f"{v.name} = {v.value()}")
print("Total profit = $", pl.value(prob.objective))
```
