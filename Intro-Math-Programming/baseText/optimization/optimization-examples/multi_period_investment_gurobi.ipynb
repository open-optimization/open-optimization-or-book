```python
# Multi-Period Capital Investment (0-1) — gurobipy
import gurobipy as gp
from gurobipy import GRB
inv = {"A":(1,10000,15000),"B":(1,20000,30000),"C":(1,15000,22000),
       "D":(2,12000,18000),"E":(2,25000,35000),"F":(2,18000,26000),
       "G":(3,15000,20000),"H":(3,10000,14000),"I":(3,20000,28000),
       "J":(4, 8000,10000),"K":(4,16000,22000),"L":(4,12000,16000)}
START=50000
m = gp.Model("Multi_Period_Investment")
x = {i: m.addVar(vtype=GRB.BINARY, name=f"x_{i}") for i in inv}
m.setObjective(gp.quicksum(p*x[i] for i,(t,c,p) in inv.items() if t==4), GRB.MAXIMIZE)
for period in (1,2,3,4):
    cost_t = gp.quicksum(c*x[i] for i,(t,c,p) in inv.items() if t==period)
    if period==1: m.addConstr(cost_t<=START,"Budget_1")
    else: m.addConstr(cost_t<=gp.quicksum(p*x[i] for i,(t,c,p) in inv.items() if t==period-1), f"Budget_{period}")
m.optimize()
if m.status==GRB.OPTIMAL:
    print("Chosen:", [i for i in inv if x[i].X>0.5]); print("Final cash =", m.ObjVal)
```
