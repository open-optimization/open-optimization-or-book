```python
import gurobipy as gp
from gurobipy import GRB

# Create a new model
m = gp.Model("diet")

# Create variables (integer)
x1 = m.addVar(name="green", vtype=GRB.INTEGER)
x2 = m.addVar(name="blue", vtype=GRB.INTEGER)
x3 = m.addVar(name="yellow", vtype=GRB.INTEGER)
x4 = m.addVar(name="red", vtype=GRB.INTEGER)

# Set objective
m.setObjective(1.25 * x1 + 1.05 * x2 + 0.85 * x3 + 0.65 * x4, GRB.MINIMIZE)

# Add constraint: Iron
m.addConstr(6 * x1 + 4 * x2 + 5 * x3 + 3 * x4 >= 20, "Iron")

# Add constraint: Vitamin B
m.addConstr(6 * x1 + 5 * x2 + 2 * x3 + 6 * x4 >= 25, "Vitamin_B")

# Add constraint: Vitamin C
m.addConstr(7 * x1 + 4 * x2 + 5 * x3 + 3 * x4 >= 30, "Vitamin_C")

# Add constraint: Vitamin D
m.addConstr(4 * x1 + 9 * x2 + 6 * x3 + 2 * x4 >= 15, "Vitamin_D")

# Optimize model
m.optimize()

# Print solution
if m.status == GRB.OPTIMAL:
    print('Optimal solution:')
    for v in m.getVars():
        print(f'{v.varName}, {v.x}')
else:
    print('No optimal solution found.')
```
