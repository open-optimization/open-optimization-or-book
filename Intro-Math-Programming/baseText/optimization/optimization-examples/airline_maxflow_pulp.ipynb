```python
# Airline Transfer Network: Maximum Flow s -> t — PuLP
import pulp as pl
cap = {("s","a"):4,("s","b"):2,("s","c"):8,("a","c"):5,("a","d"):2,("b","c"):6,
       ("b","e"):9,("c","d"):1,("c","e"):3,("c","t"):4,("d","e"):7,("e","t"):5}
V = {"s","a","b","c","d","e","t"}; s,t = "s","t"
prob = pl.LpProblem("Max_Flow", pl.LpMaximize)
x = {a: pl.LpVariable(f"x_{a[0]}_{a[1]}", lowBound=0, upBound=u) for a,u in cap.items()}
prob += pl.lpSum(x[a] for a in cap if a[0]==s), "Flow_out_of_s"
for k in V - {s,t}:
    prob += (pl.lpSum(x[a] for a in cap if a[0]==k) ==
             pl.lpSum(x[a] for a in cap if a[1]==k)), f"Balance_{k}"
prob.solve()
print("Status:", pl.LpStatus[prob.status])
print("Max passengers (flow) =", pl.value(prob.objective))
for a in cap: print(f"{a[0]}->{a[1]}: {x[a].value():g}")
```
