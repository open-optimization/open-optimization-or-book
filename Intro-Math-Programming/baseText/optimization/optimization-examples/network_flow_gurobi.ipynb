```python
# Minimum-Cost Network Flow (2 warehouses -> 2 stores) — gurobipy
import gurobipy as gp
from gurobipy import GRB
arcs = {("W1","S1"):(4,15), ("W1","S2"):(6,10), ("W2","S1"):(5,20), ("W2","S2"):(2,20)}
supply = {"W1":20, "W2":30}; demand = {"S1":25, "S2":25}
m = gp.Model("Min_Cost_Flow")
x = {a: m.addVar(lb=0, ub=cap, name=f"x_{a[0]}_{a[1]}") for a,(c,cap) in arcs.items()}
m.setObjective(gp.quicksum(c*x[a] for a,(c,cap) in arcs.items()), GRB.MINIMIZE)
for w_ in supply: m.addConstr(gp.quicksum(x[a] for a in arcs if a[0]==w_) <= supply[w_], f"Supply_{w_}")
for s_ in demand: m.addConstr(gp.quicksum(x[a] for a in arcs if a[1]==s_) == demand[s_], f"Demand_{s_}")
m.optimize()
if m.status==GRB.OPTIMAL:
    for a in arcs: print(f"{a[0]}->{a[1]}: {x[a].X:g}")
    print("Total cost =", m.ObjVal)
```
