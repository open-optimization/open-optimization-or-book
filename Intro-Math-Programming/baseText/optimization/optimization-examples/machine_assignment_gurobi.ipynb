# Machine Assignment — gurobipy
Companion to the Machine Assignment example in the compact-notation modeling chapter.
Optimal assignment: Machine 0 -> Job 1, Machine 1 -> Job 0, Machine 2 -> Job 2, Machine 3 -> Job 3, total cost 13.


```python
# Machine Assignment (4 machines, 4 jobs) — gurobipy
import gurobipy as gp
from gurobipy import GRB

C = [[9, 2, 7, 8],
     [6, 4, 3, 7],
     [5, 8, 1, 8],
     [7, 6, 9, 4]]   # C[i][j] = cost of machine i doing job j
I = range(4); J = range(4)

m = gp.Model("MachineAssignment")
x = m.addVars(I, J, vtype=GRB.BINARY, name="x")

m.setObjective(gp.quicksum(C[i][j] * x[i, j] for i in I for j in J), GRB.MINIMIZE)
m.addConstrs((x.sum("*", j) == 1 for j in J), name="job_assigned")
m.addConstrs((x.sum(i, "*") == 1 for i in I), name="machine_assigned")

m.optimize()
for i in I:
    for j in J:
        if x[i, j].X > 0.5:
            print(f"Machine {i} -> Job {j}  (cost {C[i][j]})")
print("Total cost:", m.ObjVal)
```
