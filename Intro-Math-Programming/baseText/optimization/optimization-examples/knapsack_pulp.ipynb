```python
# 0/1 Knapsack (W=15) — PuLP
import pulp as pl
w = [12, 2, 1, 1, 4]     # weights (kg)
v = [4,  2, 2, 1, 10]    # values ($)
W = 15
prob = pl.LpProblem("Knapsack", pl.LpMaximize)
x = [pl.LpVariable(f"x{i+1}", cat="Binary") for i in range(len(w))]
prob += pl.lpSum(v[i]*x[i] for i in range(len(w))), "Total_value"
prob += pl.lpSum(w[i]*x[i] for i in range(len(w))) <= W, "Capacity"
prob.solve()
print("Status:", pl.LpStatus[prob.status])
print("Take items:", [i+1 for i in range(len(w)) if x[i].value()==1])
print("Total value =", pl.value(prob.objective))
```
