# Machine Assignment — PuLP
Companion to the Machine Assignment example in the compact-notation modeling chapter.


```python
# Machine Assignment (4 machines, 4 jobs) — PuLP
# Assign one job to each machine (and one machine to each job) minimizing total cost.
import pulp as pl

C = [[9, 2, 7, 8],
     [6, 4, 3, 7],
     [5, 8, 1, 8],
     [7, 6, 9, 4]]   # C[i][j] = cost of machine i doing job j
I = range(4); J = range(4)

prob = pl.LpProblem("MachineAssignment", pl.LpMinimize)
x = pl.LpVariable.dicts("x", (I, J), cat="Binary")

prob += pl.lpSum(C[i][j] * x[i][j] for i in I for j in J), "Total_cost"
for j in J:
    prob += pl.lpSum(x[i][j] for i in I) == 1, f"job_{j}_assigned"
for i in I:
    prob += pl.lpSum(x[i][j] for j in J) == 1, f"machine_{i}_assigned"

prob.solve(pl.PULP_CBC_CMD(msg=0))
print("Status:", pl.LpStatus[prob.status])
for i in I:
    for j in J:
        if x[i][j].value() > 0.5:
            print(f"Machine {i} -> Job {j}  (cost {C[i][j]})")
print("Total cost:", pl.value(prob.objective))
```
