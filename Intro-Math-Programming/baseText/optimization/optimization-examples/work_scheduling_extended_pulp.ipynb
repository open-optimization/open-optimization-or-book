```python
import pulp

# Define the problem
problem = pulp.LpProblem("LP_Burger_Extended", pulp.LpMinimize)

# Decision Variables
x = pulp.LpVariable.dicts("x", range(7), lowBound=0, cat='Integer')  # Full-time workers
y = pulp.LpVariable.dicts("y", range(7), lowBound=0, cat='Integer')  # Part-time workers

# Objective Function
problem += 5*pulp.lpSum(x[i] for i in range(7)) + 2*pulp.lpSum(y[i] for i in range(7)), "TotalCost"

# Constraints for full-time workers
for i in range(7):
    problem += x[i] + x[(i+3) % 7] + x[(i+4) % 7] + x[(i+5) % 7] + x[(i+6) % 7] + y[i] + y[(i-1) % 7] >= [6, 4, 5, 4, 3, 7, 7][i], f"Day_{i+1}"

# Part-time workers constraint
problem += pulp.lpSum(y[i] for i in range(7)) <= 2, "PartTimeLimit"

# Solve the problem
problem.solve()

# Print the solution
print("Status:", pulp.LpStatus[problem.status])
print("Minimum number of full-time and part-time employees required:", pulp.value(problem.objective))
for i in range(7):
    print(f"Full-time start on day {i+1}: {x[i].value()}")
    print(f"Part-time start on day {i+1}: {y[i].value()}")
```
