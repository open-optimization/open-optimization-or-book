```python
# Minimizing an absolute value via |t| = max(t, -t)  — gurobipy
import gurobipy as gp
from gurobipy import GRB

pts = [2, 5, 9]
m = gp.Model("Min_Absolute_Deviation")
x = m.addVar(lb=-GRB.INFINITY, name="x")
z = m.addVars(len(pts), lb=0, name="z")
m.setObjective(gp.quicksum(z[i] for i in range(len(pts))), GRB.MINIMIZE)
for i, p in enumerate(pts):
    m.addConstr(x - p <= z[i], f"upper_{i}")
    m.addConstr(p - x <= z[i], f"lower_{i}")
m.optimize()
if m.status == GRB.OPTIMAL:
    print("x =", x.X, " (the median of the points)")
    print("Total absolute deviation =", m.ObjVal)
```
