```python
from gurobipy import GRB, Model

# Create a new model
m = Model("LP_Burger_Extended")

# Decision variables
days = range(7)  # Monday to Sunday
x = m.addVars(days, name="x", vtype=GRB.INTEGER)  # Full-time workers
y = m.addVars(days, name="y", vtype=GRB.INTEGER)  # Part-time workers

# Objective function
m.setObjective(5*sum(x[i] for i in days) + 2*sum(y[i] for i in days), GRB.MINIMIZE)

# Constraints for full-time workers
m.addConstrs((x[i] + x[(i+3) % 7] + x[(i+4) % 7] + x[(i+5) % 7] + x[(i+6) % 7] 
             + y[i] + y[(i-1) % 7] >= [6, 4, 5, 4, 3, 7, 7][i] for i in days), "Staffing")

# Part-time workers constraint
m.addConstr(sum(y[i] for i in days) <= 2, "PartTimeLimit")

# Solve model
m.optimize()

# Print solution
if m.status == GRB.OPTIMAL:
    print("Minimum number of full-time and part-time employees required:", m.objVal)
    for i in days:
        print(f"Full-time start on day {i+1}: {x[i].x}")
        print(f"Part-time start on day {i+1}: {y[i].x}")
else:
    print("No optimal solution found.")
```
