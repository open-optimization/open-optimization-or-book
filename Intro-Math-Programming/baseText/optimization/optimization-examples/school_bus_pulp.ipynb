```python
# School Bus Routing (assignment) — PuLP
import pulp as pl
buses   = [1,2,3,4,5]
schools = ["A","B","C","D","E"]
cost = {  # cost[bus][school]
 1:{"A":50,"B":80,"C":60,"D":90,"E":100},
 2:{"A":60,"B":85,"C":55,"D":70,"E":110},
 3:{"A":75,"B":65,"C":50,"D":85,"E":90},
 4:{"A":70,"B":90,"C":55,"D":80,"E":95},
 5:{"A":80,"B":70,"C":60,"D":75,"E":85}}
prob = pl.LpProblem("School_Bus_Assignment", pl.LpMinimize)
x = pl.LpVariable.dicts("x",(buses,schools),cat="Binary")
prob += pl.lpSum(cost[j][i]*x[j][i] for j in buses for i in schools), "Total_fuel_cost"
for j in buses:   prob += pl.lpSum(x[j][i] for i in schools)==1, f"Bus_{j}"
for i in schools: prob += pl.lpSum(x[j][i] for j in buses)==1, f"School_{i}"
prob.solve()
print("Status:", pl.LpStatus[prob.status])
for j in buses:
    for i in schools:
        if x[j][i].value()==1: print(f"Bus {j} -> School {i} (cost {cost[j][i]})")
print("Total fuel cost =", pl.value(prob.objective))
```
