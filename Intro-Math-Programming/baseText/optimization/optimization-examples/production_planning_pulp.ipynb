```python
# Production Planning (two-day inventory) — PuLP
# min 10 x1 + 16 x2 + 5 s1 + 5 s2
# s0 + x1 = 10 + s1 ;  s1 + x2 = 4 + s2 ;  x,s >= 0 ;  s0 = 0 (initial inventory)
import pulp as pl

prob = pl.LpProblem("Production_Planning", pl.LpMinimize)
s0 = 0  # initial inventory

x1 = pl.LpVariable("x1_produce_day1", lowBound=0)
x2 = pl.LpVariable("x2_produce_day2", lowBound=0)
s1 = pl.LpVariable("s1_inventory_day1", lowBound=0)
s2 = pl.LpVariable("s2_inventory_day2", lowBound=0)

prob += 10*x1 + 16*x2 + 5*s1 + 5*s2, "Total_Cost"
prob += s0 + x1 == 10 + s1, "Day1_balance"
prob += s1 + x2 == 4 + s2,  "Day2_balance"

prob.solve()
print("Status:", pl.LpStatus[prob.status])
for v in prob.variables():
    print(f"{v.name} = {v.value()}")
print("Total cost =", pl.value(prob.objective))
```
