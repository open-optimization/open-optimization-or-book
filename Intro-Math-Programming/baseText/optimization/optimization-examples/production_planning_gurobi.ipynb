```python
# Production Planning (two-day inventory) — gurobipy
import gurobipy as gp
from gurobipy import GRB

m = gp.Model("Production_Planning")
s0 = 0  # initial inventory

x1 = m.addVar(lb=0, name="x1_produce_day1")
x2 = m.addVar(lb=0, name="x2_produce_day2")
s1 = m.addVar(lb=0, name="s1_inventory_day1")
s2 = m.addVar(lb=0, name="s2_inventory_day2")

m.setObjective(10*x1 + 16*x2 + 5*s1 + 5*s2, GRB.MINIMIZE)
m.addConstr(s0 + x1 == 10 + s1, "Day1_balance")
m.addConstr(s1 + x2 == 4 + s2,  "Day2_balance")

m.optimize()
if m.status == GRB.OPTIMAL:
    for v in m.getVars():
        print(f"{v.VarName} = {v.X}")
    print("Total cost =", m.ObjVal)
```
