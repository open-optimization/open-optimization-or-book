```python
# School Bus Routing (assignment) — gurobipy
import gurobipy as gp
from gurobipy import GRB
buses=[1,2,3,4,5]; schools=["A","B","C","D","E"]
cost={1:{"A":50,"B":80,"C":60,"D":90,"E":100},2:{"A":60,"B":85,"C":55,"D":70,"E":110},
 3:{"A":75,"B":65,"C":50,"D":85,"E":90},4:{"A":70,"B":90,"C":55,"D":80,"E":95},
 5:{"A":80,"B":70,"C":60,"D":75,"E":85}}
m=gp.Model("School_Bus_Assignment")
x=m.addVars(buses,schools,vtype=GRB.BINARY,name="x")
m.setObjective(gp.quicksum(cost[j][i]*x[j,i] for j in buses for i in schools),GRB.MINIMIZE)
m.addConstrs((gp.quicksum(x[j,i] for i in schools)==1 for j in buses),"Bus")
m.addConstrs((gp.quicksum(x[j,i] for j in buses)==1 for i in schools),"School")
m.optimize()
if m.status==GRB.OPTIMAL:
    for j in buses:
        for i in schools:
            if x[j,i].X>0.5: print(f"Bus {j} -> School {i} (cost {cost[j][i]})")
    print("Total fuel cost =", m.ObjVal)
```
