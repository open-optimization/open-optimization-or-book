```python
from pulp import *

# Problem data
projects = ['A', 'B', 'C', 'D', 'E']
investment_required = {'A': 20000, 'B': 30000, 'C': 50000, 'D': 10000, 'E': 40000}
projected_return = {'A': 30000, 'B': 40000, 'C': 75000, 'D': 15000, 'E': 60000}

# Create the LP object, set up as a maximization problem
prob = LpProblem("Capital_Allocation", LpMaximize)

# Decision variables
x = LpVariable.dicts("Project", projects, 0, 1, LpBinary)

# Objective function
prob += lpSum([projected_return[i] * x[i] for i in projects]), "Total_Projected_Return"

# Constraints
prob += lpSum([investment_required[i] * x[i] for i in projects]) <= 100000, "Total_Investment"

# Solve the problem
prob.solve()

# Print the results
print("Status:", LpStatus[prob.status])
for v in prob.variables():
    print(v.name, "=", v.varValue)

print("Total Projected Return =", value(prob.objective))
```
