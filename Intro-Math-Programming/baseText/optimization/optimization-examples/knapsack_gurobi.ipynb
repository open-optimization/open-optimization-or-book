```python
# 0/1 Knapsack (W=15) — gurobipy
import gurobipy as gp
from gurobipy import GRB
w=[12,2,1,1,4]; v=[4,2,2,1,10]; W=15
m=gp.Model("Knapsack")
x=m.addVars(len(w),vtype=GRB.BINARY,name="x")
m.setObjective(gp.quicksum(v[i]*x[i] for i in range(len(w))),GRB.MAXIMIZE)
m.addConstr(gp.quicksum(w[i]*x[i] for i in range(len(w)))<=W,"Capacity")
m.optimize()
if m.status==GRB.OPTIMAL:
    print("Take items:",[i+1 for i in range(len(w)) if x[i].X>0.5]); print("Total value =",m.ObjVal)
```
