```python
# Minimum-Cost Network Flow (2 warehouses -> 2 stores) — PuLP
import pulp as pl
# arc: (cost, capacity);  supplies and demands
arcs = {("W1","S1"):(4,15), ("W1","S2"):(6,10), ("W2","S1"):(5,20), ("W2","S2"):(2,20)}
supply = {"W1":20, "W2":30}
demand = {"S1":25, "S2":25}

prob = pl.LpProblem("Min_Cost_Flow", pl.LpMinimize)
x = {a: pl.LpVariable(f"x_{a[0]}_{a[1]}", lowBound=0, upBound=cap) for a,(c,cap) in arcs.items()}
prob += pl.lpSum(c*x[a] for a,(c,cap) in arcs.items()), "Total_Cost"
for w_ in supply:
    prob += pl.lpSum(x[a] for a in arcs if a[0]==w_) <= supply[w_], f"Supply_{w_}"
for s_ in demand:
    prob += pl.lpSum(x[a] for a in arcs if a[1]==s_) == demand[s_], f"Demand_{s_}"
prob.solve()
print("Status:", pl.LpStatus[prob.status])
for a in arcs: print(f"{a[0]}->{a[1]}: {x[a].value():g}")
print("Total cost =", pl.value(prob.objective))
```
