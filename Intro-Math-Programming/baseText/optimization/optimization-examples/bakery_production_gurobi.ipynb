```python
# Bakery Production — gurobipy
import gurobipy as gp
from gurobipy import GRB

m = gp.Model("Bakery_Production")
c  = m.addVar(vtype=GRB.INTEGER, lb=0, name="cakes")
k  = m.addVar(vtype=GRB.INTEGER, lb=0, name="cookies")
mu = m.addVar(vtype=GRB.INTEGER, lb=0, name="muffins")

m.setObjective(6*c + 1.5*k + 2.5*mu, GRB.MAXIMIZE)
m.addConstr(500*c + 100*k + 200*mu <= 4800, "Flour_g")
m.addConstr(200*c +  50*k +  80*mu <= 2060, "Sugar_g")
m.addConstr( 60*c +  10*k +  20*mu <= 480,  "Baking_minutes")

m.optimize()
if m.status == GRB.OPTIMAL:
    for v in m.getVars():
        print(f"{v.VarName} = {v.X}")
    print("Total profit = $", m.ObjVal)
```
