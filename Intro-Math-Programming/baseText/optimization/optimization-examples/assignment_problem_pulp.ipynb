```python
# Assignment Problem (Hiring for tasks) — PuLP
import pulp as pl

cost = {(1,1):40,(1,2):47,(1,3):80,
        (2,1):72,(2,2):36,(2,3):58,
        (3,1):24,(3,2):61,(3,3):71}
P = [1,2,3]; T = [1,2,3]

prob = pl.LpProblem("Assignment", pl.LpMinimize)
x = pl.LpVariable.dicts("x", (P,T), cat="Binary")

prob += pl.lpSum(cost[(i,j)]*x[i][j] for i in P for j in T), "Total_Cost"
for i in P:
    prob += pl.lpSum(x[i][j] for j in T) == 1, f"Person_{i}"
for j in T:
    prob += pl.lpSum(x[i][j] for i in P) == 1, f"Task_{j}"

prob.solve()
print("Status:", pl.LpStatus[prob.status])
for i in P:
    for j in T:
        if x[i][j].value() == 1:
            print(f"Person {i} -> Task {j}  (cost {cost[(i,j)]})")
print("Total cost = $", pl.value(prob.objective))
```
