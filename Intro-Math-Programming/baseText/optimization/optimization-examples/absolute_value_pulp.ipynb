```python
# Minimizing an absolute value via |t| = max(t, -t)  — PuLP
# Worked instance: find x minimizing the total absolute distance to the points 2, 5, 9.
#   min sum_i |x - p_i|   ->   min sum_i z_i  s.t.  x - p_i <= z_i,  p_i - x <= z_i.
import pulp as pl

pts = [2, 5, 9]
prob = pl.LpProblem("Min_Absolute_Deviation", pl.LpMinimize)
x = pl.LpVariable("x")                                   # free variable
z = [pl.LpVariable(f"z{i}", lowBound=0) for i in range(len(pts))]

prob += pl.lpSum(z), "Total_absolute_deviation"
for i, p in enumerate(pts):
    prob += x - p <= z[i], f"upper_{i}"
    prob += p - x <= z[i], f"lower_{i}"

prob.solve()
print("Status:", pl.LpStatus[prob.status])
print("x =", x.value(), " (the median of the points)")
print("Total absolute deviation =", pl.value(prob.objective))
```
