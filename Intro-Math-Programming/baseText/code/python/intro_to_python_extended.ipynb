## 🐍 Python Basics

Python is a high-level, interpreted programming language known for its readability and versatility.

### Variables and Data Types
Python is dynamically typed, so you don’t need to declare variable types explicitly.

- **Integers**: Whole numbers (e.g., 10, -5)
- **Floats**: Decimal numbers (e.g., 3.14, -2.7)
- **Strings**: Sequences of characters (e.g., "Hello", "Alice")
- **Booleans**: True or False


```python
# Basic variable assignment
x = 10           # Integer
y = 3.14         # Float
name = "Alice"   # String
is_happy = True  # Boolean

print(x, y, name, is_happy)
```

### Basic Operations

Python supports common arithmetic operations:
- `+` (addition)
- `-` (subtraction)
- `*` (multiplication)
- `/` (division)
- `**` (exponentiation)
- `//` (floor division)
- `%` (modulus)



```python
# Demonstration of basic arithmetic
a = 15
b = 4

print("Addition:", a + b)
print("Subtraction:", a - b)
print("Multiplication:", a * b)
print("Division:", a / b)
print("Floor division:", a // b)
print("Modulus:", a % b)
print("Exponentiation:", a ** b)
```

### Collections: Lists, Tuples, Dictionaries, and Sets

These are fundamental data structures in Python:

- **Lists** are mutable sequences (e.g., `[1, 2, 3]`).
- **Tuples** are immutable sequences (e.g., `(10, 20, 30)`).
- **Dictionaries** store key-value pairs (e.g., `{"apple": 2, "banana": 3}`).
- **Sets** store unique elements (e.g., `{1, 2, 3}`).


```python
my_list = [1, 2, 3, 4]              # List
my_tuple = (10, 20, 30)              # Tuple
my_dict = {"apple": 2, "banana": 3}  # Dictionary
my_set = {1, 2, 2, 3, 4}             # Set (duplicates are removed)

print("List:", my_list)
print("Tuple:", my_tuple)
print("Dictionary:", my_dict)
print("Set:", my_set)

# List operations
my_list.append(5)
print("\nAfter appending 5 to my_list:", my_list)

# Dictionary operations
my_dict["cherry"] = 5
print("\nAfter adding cherry to my_dict:", my_dict)
```

### Indexing and Slicing

Lists, tuples, and strings support indexing (to access a single element) and slicing (to access a subrange of elements).

list_or_string[start:end:step]
- `start`: The starting index (default is 0)
- `end`: The ending index (exclusive, default is length of the sequence)
- `step`: The step size (default is 1)


```python
# Indexing and slicing
colors = ["red", "green", "blue", "yellow", "purple"]
print("colors[0]:", colors[0])
print("colors[1:3]:", colors[1:3])
print("colors[:3]:", colors[:3])
print("colors[::2]:", colors[::2])
```

### Control Flow

Python uses indentation to define blocks. Key statements include:
- **if**, **elif**, **else** for conditional branching
- **for** loops for iteration over sequences
- **while** loops for repeated execution as long as a condition holds


```python
# Using a for loop to iterate over a dictionary
for fruit in my_dict:
    print(f"{fruit} has {my_dict[fruit]} items.")

# Example of if-else
num = 7
if num % 2 == 0:
    print(f"{num} is even.")
else:
    print(f"{num} is odd.")
```

### Functions

Use `def` to define reusable blocks of code. Functions can take parameters and return values.


```python
def square(x):
    return x * x

val = 5
print(f"The square of {val} is {square(val)}.")
```

## 🔢 NumPy: Numerical Computing in Python

NumPy is the fundamental package for numerical computation in Python. It provides:
- A powerful N-dimensional array object
- Functions for performing vectorized operations
- Tools for linear algebra, random number generation, and more


```python
import numpy as np

# Create a 1D array
a = np.array([1, 2, 3, 4, 5])
print("Array a:", a)

print("Mean:", np.mean(a))
print("Standard Deviation:", np.std(a))
print("Element-wise multiplication:", a * 2)
```

### Multi-Dimensional Arrays
NumPy supports multi-dimensional arrays (matrices, tensors). Let's create a 2D array (matrix) and explore some operations.


```python
# 2D array (matrix)
b = np.array([[1, 2],
              [3, 4]])
print("Matrix b:\n", b)

# Transpose
print("\nTranspose of b:\n", b.T)

# Shape
print("\nShape of b:", b.shape)

# Reshaping
c = np.array([1,2,3,4,5,6])
c_reshaped = c.reshape((2, 3))
print("\nOriginal c:", c)
print("Reshaped c to (2,3):\n", c_reshaped)
```

### Random Data Generation

NumPy provides a convenient way to generate random data, which is often useful for simulations or testing.


```python
# Generating random data
np.random.seed(42)  # for reproducibility

rand_array = np.random.rand(3, 3)  # 3x3 array of random numbers in [0,1)
print("Random 3x3 array:\n", rand_array)

rand_ints = np.random.randint(0, 10, size=(2, 5))  # 2x5 array of random integers in [0,10)
print("\nRandom 2x5 array of integers:\n", rand_ints)
```

## 🐼 Pandas: Data Manipulation Made Easy

Pandas is a powerful library for data analysis and manipulation. It introduces two primary data structures:

- **Series**: A 1D labeled array  
- **DataFrame**: A 2D labeled table (like a spreadsheet)

Pandas makes it easy to load, clean, and analyze data from various sources, including CSV files, Excel files, SQL databases, and more.


```python
import pandas as pd

# Creating a simple DataFrame from a dictionary
data = {
    "Name": ["Alice", "Bob", "Charlie"],
    "Age": [25, 30, 35],
    "Score": [85.0, 92.5, 88.0]
}
df = pd.DataFrame(data)
df
```

### Basic DataFrame Operations

Pandas DataFrames come with many built-in methods and properties that allow for quick data manipulation.


```python
# Check the average score
print("Average Score:", df["Score"].mean())

# Sort by age
print("\nSorted by Age:")
print(df.sort_values("Age"))

# Accessing columns
print("\nAccessing the 'Name' column:")
print(df["Name"])

# Filtering rows
print("\nFiltering rows where Age > 25:")
print(df[df["Age"] > 25])
```

### Grouping and Aggregation

Grouping data allows you to split the data into groups based on some criteria, apply a function to each group, and then combine the results.


```python
# Create a slightly bigger DataFrame for demonstration
data_extended = {
    "Name": ["Alice", "Bob", "Charlie", "Alice", "Bob", "Charlie"],
    "Category": ["A", "A", "B", "B", "A", "B"],
    "Value": [10, 20, 5, 15, 30, 25]
}
df_ext = pd.DataFrame(data_extended)
print(df_ext)

# Group by Category and sum the Value
grouped = df_ext.groupby("Category")["Value"].sum()
print("\nSum of Value by Category:")
print(grouped)
```

## 📁 Importing Real Data

Let's import a real-world dataset from a CSV file available online. We'll use air travel data.

> **Note**: If you run into any issues loading data from a URL (e.g., due to network restrictions), you can download the file and load it locally instead.


```python
# Load dataset from URL
url = "https://people.sc.fsu.edu/~jburkardt/data/csv/airtravel.csv"
df_air = pd.read_csv(url)
df_air.head()
```

```python
# Checking structure
df_air.info()
```

### Slicing and Filtering DataFrames

In Pandas, you can filter a DataFrame by applying a boolean condition to one of its columns.
For example, if you only want rows where a certain numeric column is larger than a threshold, you can do:

`df[df["column_name"] > some_value]`

You can combine multiple conditions using the bitwise operators `&` (AND) and `|` (OR) with parentheses:

`df[(df["colA"] > val1) & (df["colB"] == val2)]`

Below, we show a practical example.


```python
# Example DataFrame
df_filter_example = pd.DataFrame({
    "Name": ["Alice", "Bob", "Charlie", "Diana", "Ethan"],
    "Age": [25, 19, 34, 42, 17],
    "Score": [85.0, 92.5, 88.0, 75.0, 95.0]
})

print("Original DataFrame:")
display(df_filter_example)

# Filter rows where Age is greater than 20
filtered_df = df_filter_example[df_filter_example["Age"] > 20]
print("\nFiltered DataFrame (Age > 20):")
display(filtered_df)
```

## 📊 Matplotlib: Basic Visualization

Matplotlib is a library for creating visualizations in Python. It offers a wide variety of plots and customization options.

### Basic Line Plot

We’ll use the air travel data to create a line plot.


```python
import matplotlib.pyplot as plt

months = df_air["Month"]
totals = df_air.iloc[:, 1:].sum(axis=1)

plt.figure(figsize=(8, 4))
plt.plot(months, totals, marker='o', linestyle='-')
plt.title("Total Air Travel Per Month")
plt.xlabel("Month")
plt.ylabel("Total Passengers")
plt.xticks(rotation=45)
plt.grid(True)
plt.tight_layout()
plt.show()
```

### Bar Plot
Bar plots are great for categorical data or to compare different categories.


```python
plt.figure(figsize=(8, 4))
plt.bar(months, totals)
plt.title("Total Air Travel Per Month (Bar Chart)")
plt.xlabel("Month")
plt.ylabel("Total Passengers")
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

### Scatter Plot
If you have two continuous variables, a scatter plot helps visualize their relationship. Below is a quick example with random data.


```python
x_rand = np.random.rand(50)
y_rand = np.random.rand(50)

plt.figure(figsize=(8, 4))
plt.scatter(x_rand, y_rand)
plt.title("Random Scatter Plot")
plt.xlabel("X values")
plt.ylabel("Y values")
plt.tight_layout()
plt.show()
```

### Histogram
Histograms help visualize the distribution of numerical data.


```python
data_for_hist = np.random.randn(1000)  # 1000 normally distributed points

plt.figure(figsize=(8, 4))
plt.hist(data_for_hist, bins=20)  # 20 bins
plt.title("Histogram of 1000 Random Values")
plt.xlabel("Value")
plt.ylabel("Frequency")
plt.tight_layout()
plt.show()
```
